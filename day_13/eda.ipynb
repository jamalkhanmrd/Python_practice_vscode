```python
df.drop(columns="deck",inplace=True )
```

```python
df.columns
```

```python
#visualize and explore
sns.countplot(data=df,x="survived")
```

```python
sns.barplot(data=df,x="survived",y="age",hue="who")
```

```python
sns.countplot(x="pclass",data=df,hue="who")
```

```python
#data distribution
sns.histplot(data=df,x="age",bins=7)
```

```python
df["age"].hist(bins=38,color="darkred",alpha=0.9)
```

```python
sns.boxplot(data=df,y="age",x="survived",hue="who")
```

```python
sns.scatterplot(x="age",y="fare",data=df,hue="pclass")
# plt.ylim(0,250)
```

```python
sns.pairplot(df,hue="survived")
```

```python
#stacked barchat /plot
df.groupby(["pclass","who"])["survived"].mean().unstack().plot(kind="bar",stacked=True)
```

```python
#swarm plot
sns.swarmplot(data=df,x="survived",y="age",hue="who")
```

```python
sns.lmplot(data=df,y="fare",x=df[df.fare<300])
```

```python
sns.lmplot(data=df[df["fare"] < 300], x="fare", y="fare",hue="who")
```

```python
# how to add a culumn and make a age group
# grouping
df["age_groups"]=pd.cut(df["age"],[0,10,18,30,50,80],labels=["kids","child","youth","middle age","old"])
```

# how to rename a column

```python
df=pd.DataFrame({'colum A':[1,2,3],'colum B':[4,5,6]})
df
```

```python
df.rename(columns={"A_colum":"colum_a","B_colum":"col_b"},inplace=True)
df
```

```python
df.columns=["Col_a","col_b"]
df
```

```python
# to replace any character
df.columns=df.columns.str.replace('_',' ')
df
```

```python
# adding prefix
df=df.add_prefix('col_a')
df
```

```python
# adding psuffix
df=df.add_suffix('baba_a')
df
```

```python
df.columns=['col_a','col_b']
df
```
